# Kepler problem.

Load data and plot $t$, $x$, $y$

```python
from pylab import *
%matplotlib inline

t = loadtxt('t.txt')
x = loadtxt('x.txt')
y = loadtxt('y.txt')
# Draw x(t) vs t 
figure()
plot(x, t, 'b')
xlabel('x(x)')
ylabel('t')
title('$x(t)$ vs $t$')
# Draw y(t) vs t 
figure()
plot(y, t, 'r')
xlabel('y(t)')
ylabel('t')
title('$y(t)$ vs $t$')
# Draw y(t) vs x(t)
figure()
plot(y, x, 'g')
xlabel('y(t)')
ylabel('x(t)')
title('$y(t)$ vs $x(t)$')
```
