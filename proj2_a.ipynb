# Linear Solver. (Vandermonde Matrix)

Load data and plot $error$ vector respect to $N$ (size of Vandermonde matrix)

```python
from pylab import *
%matplotlib inline
N = loadtxt('N.txt')
error = loadtxt('error.txt')
figure()
loglog(N, error, 'r')
xlabel('N')
ylabel('error')
title('2-norm of $x$ values error respect to $N$')
show()
```

Now load data and plot $residula$ vector with values for each $N$

```python
from pylab import *
%matplotlib inline
N = loadtxt('N.txt')
residula = loadtxt('residula.txt')
figure()
loglog(N, residula, 'b')
xlabel('N')
ylabel('residula')
title('2-norm of $b$ values residula respect to $N$')
show()
```

As we can see both vectors: $error$ and $residula$ continue to rise when $N$ values increase. Float point error probably answer to this pictures. Because of more float operations -> greater error.
