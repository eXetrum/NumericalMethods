# Taylor series script

Load  and plot data generated with c++, vectors: n, h, r, R. First lets load $f(n, r)$, $f(n, R)$

```python
from pylab import *
%matplotlib inline
n = loadtxt('n.txt')
r = loadtxt('r.txt')
R = loadtxt('R_.txt')
figure()
semilogy(n, r, 'b')
hold
semilogy(n, R, 'r--')
xlabel('n')
ylabel('values')
title('Errors in a Forward Finite Difference Approximation')
legend(['r versus n','R versus n'])
show()
```

As we can see the upper bound $R$ is decreasing with the increasing value of $n$.  Value of relative error $r$ is decreasing at the beginning and reaching it's mininum about $n=27$ but later it starts increasing. And after $n=27$ both upper bound $R$ and relative errors $r$ is increasing with the increasing value of $n$. When both values decreasing they moves nearly, but when they  increase - only upper bound $R$ values go almost linear, relative error $r$ looks as piecewise function.
Now load data and plot  $f(h,r)$, $f(h,R)$

```python
from pylab import *
%matplotlib inline
h = loadtxt('h.txt')
r = loadtxt('r.txt')
R = loadtxt('R_.txt')
figure()
loglog(h, r, 'b')
hold
loglog(h, R, 'r--')
xlabel('h')
ylabel('Error')
title('Errors in a Forward Finite Difference Approximation')
legend(['r versus h','R versus h'])
show()
```

We can see the same on this plot. If h is decreasing (so when n in increasing) upper bound continue to rise when relative error at first is decreasing but later it starts to increase. Lowest error occured near $h=10^{-9}$

```python
from pylab import *
%matplotlib inline
n = loadtxt('n.txt')
h = loadtxt('h.txt')
r = loadtxt('r.txt')
R = loadtxt('R_.txt')
relToN  = {ni : ri for (ni, ri) in zip(n, r) }
nToStep = {ni : hi for (ni, hi) in zip(n, h) }
bestN = min(relToN.iterkeys(), key=lambda k: relToN[k])

print("Lowest relative error r=%e occured when approximation step h=%e, this can be reached with n=%f" %(relToN[bestN], nToStep[bestN], bestN))
```

We get the best approximation for $n = 27$, because then error is the lowest. The reason for that behavior is that at first value of counted difference is a big approximation so the difference is also big. Then when $n > 27$ precision of counted difference was to big to compute so the value of related error begun to increase.
